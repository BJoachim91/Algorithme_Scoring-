Contexte

Les cartes de crédit sont une méthode de contrôle des risques courante dans le secteur financier. Elle utilise les informations et les données personnelles soumises par les demandeurs de carte de crédit pour prédire la probabilité de futurs défauts de paiement et d'emprunts par carte de crédit. La banque est en mesure de décider si elle émet une carte de crédit au demandeur. Les scores de crédit permettent de quantifier objectivement l'ampleur du risque.
 
En général, les cartes de crédit sont basées sur des données historiques. Lorsqu'elles sont confrontées à de grandes fluctuations économiques. Les modèles antérieurs peuvent perdre leur pouvoir prédictif initial. Le modèle logistique est une méthode courante de notation de crédit. Parce que le modèle logistique convient aux tâches de classification binaire et peut calculer les coefficients de chaque caractéristique. Afin de faciliter la compréhension et l'exploitation, la carte de pointage multipliera le coefficient de régression logistique par une certaine valeur (comme 100) et l'arrondira.
 
Actuellement, avec le développement des algorithmes d'apprentissage automatique. Des méthodes plus prédictives telles que Boosting, Random Forest et Support Vector Machines ont été introduites dans le scoring des cartes de crédit. Cependant, ces méthodes manquent souvent de transparence. Il peut être difficile de fournir aux clients et aux régulateurs la raison du rejet ou de l'acceptation.
